```python
import pandas as pd
```

```python
df_books = pd.read_csv('./bestsellers-with-categories.csv',sep=",",header=0)
df_books.head(2)
```

```python
mayor_2016 = df_books['Year'] > 2016
```

```python
df_books[mayor_2016]
```

```python
genre_fiction = df_books['Genre']=='Fiction'
```

```python
df_books[mayor_2016 & genre_fiction]
```

```python
df_books[~mayor_2016 & ~genre_fiction]
```

```python
df_books[(df_books['Year'] > 2016) & (df_books['Genre']=='Fiction')]
```

```python
df_books[df_books['Author'].str.contains('Davis')]
```

```python
df_books[df_books['Year'].isin([2016,2017])]
```

```python
df_books[df_books['Year'].between(2016,2017)]
```

```python
df_books[df_books['Year'].eq(2016)]
```
