Agregar o eliminar dimensiones

```python
vector = np.array([1,5,9], ndmin=10)
print(vector)
vector.ndim
```

```python
expand = np.expand_dims(np.array([1,5,9]), axis=0)
print(expand)
expand.ndim
```

```python
print(vector, vector.ndim)
vector_2 = np.squeeze(vector)
print(vector_2, vector_2.ndim)
```
